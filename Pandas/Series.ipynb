### Series  
py 1D labeled array which can hold any data type
(Single column in a spreadsheet)

Series via list ( default index from zero unless specified)

```python
data= [100,102, 103, True] #list
series = pd.Series(data, index=["a","b","c","d"])
series
```

Location by label

```python
#return the value at the label
print(series.loc["b"])
```

```python
#modifying using the loc operator
series.loc["d"]=1000
series
```

Location by integer pos

```python
series.iloc[3]
```

Filtering by values

```python
series[series>=101]
```

Series via dictionaries(takes the key from dictionary as the index for the series)

```python
calories = {'Day1': 1750, 'Day2' : 2100, 'Day3': 1700}
series= pd.Series(calories)
series
```

```python
print(series.loc["Day1"])
```

```python
series.loc['Day3']+=500
series
```

Filter by values


```python
print(series[series>= 2000])
```
