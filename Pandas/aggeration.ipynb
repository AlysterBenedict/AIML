```python
#aggregate functions = reduces a set of values into a single summary value 
# used to summarize and analyze data often used with the groupby() function
import pandas as pd
df= pd.read_csv("../Datasets/pokemon.csv")

print(f" averages:\n {df.mean(numeric_only=True)}")
```

```python
print(f" min:\n {df.min(numeric_only=True)}")
```

```python
print(f" max:\n {df.max(numeric_only=True)}")
```

```python
#count : number of values in a column
print(df.count())
```

Aggregate functions for a single column

```python
print(df["Height"].mean())
print(df["Height"].count())
```

Group By function

```python
group = df.groupby("Type1")
print(group)
print(group["Height"].count())
```
