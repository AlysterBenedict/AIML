```python
#Filtering = keeping the rows that match a condition
import pandas as pd
df= pd.read_csv("../Datasets/pokemon.csv")
tall_pokemon=df[df["Height"]>=2]  #new dataframe with the matching condition
print(tall_pokemon)
```

```python
water_pokemon = df[(df["Type1"]== "Water") | (df["Type2"]== "Water")]
print(water_pokemon)
```

```python
legendary = df[(df["Legendary"] == True) & (df["Type1"] == "Fire")]
print(legendary)
```
