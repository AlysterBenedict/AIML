The process of fixing/removing: incomplete, incorrect, irrevalent data

```python
import pandas as pd
df=pd.read_csv("../Datasets/pokemon.csv")
```

```python
#clear columns
df= df.drop(columns=["Legendary"])
df
```

```python
#handling missing data
df=df.dropna(subset=["Type2"])
df
```

```python
df=pd.read_csv("../Datasets/pokemon.csv")
df=df.fillna({"Type2" : "None"})
df
```

```python
df=pd.read_csv("../Datasets/pokemon.csv")
df
```

```python
df=pd.read_csv("../Datasets/pokemon.csv")
group=df.groupby("Type2")
value=group["No"].count().idxmax()

df.fillna({"Type2" : value})
```

```python
#Standardize text
df["Name"]= df["Name"].str.lower()
print(df)
```

```python
#Fix datatypes
df["Legendary"]= df["Legendary"].astype(bool)
df
```
