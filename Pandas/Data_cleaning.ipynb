```python
df=pd.read_csv("../Datasets/pokemon.csv")
group=df.groupby("Type2")
value=group["No"].count().idxmax()

df.fillna({"Type2" : value})
```

```python
df=pd.read_csv("../Datasets/pokemon.csv")
df
```

```python
#remove duplicate values
df=df.drop_duplicates()
df
```

```python
# Calculate the minimum height
min_height = df["Height"].min()

# Fill missing Height values with the minimum
df["Height"] = df["Height"].fillna(min_height)
df
```

```python
df = df.fillna({
    "Weight": df["Weight"].mean(),
    "Height": df["Height"].median(),
    "Type2": "None" # Combining your existing logic
})
df
```
