```python
#Selection by Row
df1=pd.read_csv("../Datasets/pokemon.csv" , index_col="Name")
print(df1.loc["Bulbasaur"])
```

```python
print(df1.iloc[1])
```

```python
#selection by row with only specified columns
print(df1.loc["Charizard", ["Height", "Weight"]])
```

```python
#selection by row with range and with only specified columns
print(df1.loc["Charizard" : "Blastoise", ["Height", "Weight"]])
```
