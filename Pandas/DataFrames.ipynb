### DataFrame
A tabular datasturcture with rows and columns(2D) similar to excel spreadsheet

```python
import pandas as pd
```

```python
data= {'Name': ['Rolex', 'Leo', "Dilli"], 'Age': [20,25,30]}

df=pd.DataFrame(data, index=['Employee1', 'Employee2', 'Employee3'])  
df
```

```python
print(df.loc['Employee2'])

print()
print(df.iloc[0])
```

Adding a new column

```python
df['Job']= ["Drug boss", "Datura mafia", 'Kaithi']
df
```

Creating a new row

```python
new_row= pd.DataFrame([{'Name': 'Napolean', 'Age': '59', 'Job': 'Police'}] , index= ['Employee4'])
df=pd.concat([df, new_row])
df
```

Creating new rows

```python
new_row= pd.DataFrame([{'Name': 'Harold', 'Age': '50', 'Job': 'Second datura boss'},{'Name': 'Antony', 'Age': '55', 'Job': 'DAS & Co boss'}], index=['Employee5', 'Employee6'] )
df=pd.concat([df, new_row])
df
```
