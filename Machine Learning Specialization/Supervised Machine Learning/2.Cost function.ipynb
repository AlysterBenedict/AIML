## Computing Cost

The term 'cost' in this assignment might be a little confusing since the data is housing cost. Here, cost is a measure of how well our model is predicting the target price of the house. The term 'price' is used for housing data.

The equation for cost with one variable is:

$$J(w,b) = \frac{1}{2m} \sum_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2$$

*(Equation 1)*

where 

$$f_{w,b}(x^{(i)}) = wx^{(i)} + b$$

*(Equation 2)*

* **$f_{w,b}(x^{(i)})$** is our prediction for example $i$ using parameters $w,b$.  
* **$(f_{w,b}(x^{(i)}) -y^{(i)})^2$** is the squared difference between the target value and the prediction.   
* These differences are summed over all the $m$ examples and divided by `2m` to produce the cost, $J(w,b)$.  

> **Note:** In lecture, summation ranges are typically from 1 to $m$, while code will range from 0 to $m-1$.

The code below calculates cost by looping over each example. In each loop:
- `f_wb`, a prediction is calculated
- the difference between the target and the prediction is calculated and squared.
- this is added to the total cost.

```python
def compute_cost(x, y, w, b): 
    """
    Computes the cost function for linear regression.
    
    Args:
      x (ndarray (m,)): Data, m examples 
      y (ndarray (m,)): target values
      w,b (scalar)    : model parameters  
    
    Returns
        total_cost (float): The cost of using w,b as the parameters for linear regression
               to fit the data points in x and y
    """
    # number of training examples
    m = x.shape[0] 
    
    cost_sum = 0 
    for i in range(m): 
        f_wb = w * x[i] + b   
        cost = (f_wb - y[i]) ** 2  
        cost_sum = cost_sum + cost  
    total_cost = (1 / (2 * m)) * cost_sum  

    return total_cost
```

 Your goal is to find a model $f_{w,b}(x) = wx + b$, with parameters $w,b$,  which will accurately predict house values given an input $x$. The cost is a measure of how accurate the model is on the training data.

The cost equation (1) above shows that if $w$ and $b$ can be selected such that the predictions $f_{w,b}(x)$ match the target data $y$, the $(f_{w,b}(x^{(i)}) - y^{(i)})^2 $ term will be zero and the cost minimized. In this simple two point example, you can achieve this!

### Larger Data Set
It is instructive to view a scenario with a few more data points. This data set includes data points that do not fall on the same line. What does that mean for the cost equation? Can we find $w$, and $b$ that will give us a cost of 0? 

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480,  430,   630, 730,])
```

In the contour plot, click on a point to select `w` and `b` to achieve the lowest cost. Use the contours to guide your selections. Note, it can take a few seconds to update the graph. 

```python
import numpy as np
import matplotlib.pyplot as plt
# from shortcuts import * # Ignore this line if not applicable

# 1. Define the cost function calculation
def compute_cost(X, y, w, b):
    m = X.shape[0]
    f_wb = w * X + b
    cost = (1 / (2 * m)) * np.sum((f_wb - y) ** 2)
    return cost

# 2. Create a grid of w and b values for the surface/contour plots
w_range = np.linspace(-100, 500, 100)
b_range = np.linspace(-250, 350, 100)
W, B = np.meshgrid(w_range, b_range)

# Calculate cost at every grid point
Z = np.zeros(W.shape)
for i in range(W.shape[0]):
    for j in range(W.shape[1]):
        Z[i, j] = compute_cost(x_train, y_train, W[i, j], B[i, j])

# 3. Setup the multi-panel figure
fig = plt.figure(figsize=(14, 6))

# Left Subplot: Contour Plot
ax1 = fig.add_subplot(1, 2, 1)
# Create contours using a logarithmic scale to handle the steep cost growth
levels = np.logspace(3, 6, 20) 
contour = ax1.contour(W, B, Z, levels=levels, cmap="terrain")
ax1.clabel(contour, inline=True, fontsize=8, fmt="%1.0f")
ax1.set_title("Cost(w,b) Contour Plot")
ax1.set_xlabel("w")
ax1.set_ylabel("b")
ax1.grid(True, alpha=0.3)

# Right Subplot: 3D Surface Plot
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
surface = ax2.plot_surface(W, B, Z, cmap="twilight_shifted", edgecolor='none', alpha=0.8)
ax2.set_title("Cost(w,b) 3D Surface")
ax2.set_xlabel("w")
ax2.set_ylabel("b")
ax2.set_zlabel("J(w,b)")

# Adjust the viewing angle to closely match the lab picture
ax2.view_init(elev=30, azim=-70)

plt.tight_layout()
plt.show()
```

Above, note the dashed lines in the left plot. These represent the portion of the cost contributed by each example in your training set. In this case, values of approximately $w=209$ and $b=2.4$ provide the lowest cost. Note that, because our training examples are not on a line, the minimum cost is not zero.

### Convex Cost surface
The fact that the cost function squares the loss ensures that the 'error surface' is convex like a soup bowl. It will always have a minimum that can be reached by following the gradient in all dimensions. In the previous plot, because the $w$ and $b$ dimensions scale differently, this is not easy to recognize. The following plot, where $w$ and $b$ are symmetric, was shown in lecture:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Define range for w and b to match the axes in the image (-20 to 20)
w_range = np.linspace(-20, 20, 100)
b_range = np.linspace(-20, 20, 100)
W, B = np.meshgrid(w_range, b_range)

# 2. Define a parabolic cost function shape J(w, b) 
# We add scaling factors to shape the bowl nicely up to J = 800
Z = 0.8 * (W**2) + 1.2 * (B**2)

# 3. Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1, projection='3d')

# 4. Plot the surface with a gradient color map and slight transparency (alpha)
# 'get_cmap("gist_earth")' or 'terrain' or 'YlOrRd' offers a similar vibe; 
# '地形' / 'terrain' gives the exact blue-center to red-rim gradient seen here.
surface = ax.plot_surface(W, B, Z, cmap="terrain_r", edgecolor='black', linewidth=0.1, alpha=0.6)

# 5. Set titles and axis labels
ax.set_title("$J(w, b)$\n[You can rotate this figure]", fontsize=14, pad=20)
ax.set_xlabel('$w$', fontsize=12)
ax.set_ylabel('$b$', fontsize=12)
ax.set_zlabel('$J(w, b)$', fontsize=12)

# 6. Explicitly set limits to match your screenshot
ax.set_xlim(-20, 20)
ax.set_ylim(-20, 20)
ax.set_zlim(0, 800)

# 7. Adjust view angle to match the perspective in your image
ax.view_init(elev=40, azim=-65)

# Adjust pane backgrounds to clean white grid lines
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False

plt.tight_layout()
plt.show()
```
